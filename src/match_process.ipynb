```python
import requests
import pandas as pd
```

```python
data_url = 'https://inf553-220b8.firebaseio.com/project/matches.json'

data = requests.get(data_url)
data_json = data.json()
```

```python
# gameDuration gameId  participants  teams
column = ['gameId','gameDuration','championId','assists','kills','damageDealtToObjectives','damageDealtToTurrets',
         'damageSelfMitigated','deaths','goldEarned','totalDamageDealtToChampions','trueDamageDealt','if_win','lane']
gameids_ls, gametime_ls, champid_ls,assist_ls, kill_ls,dm_t_ob_ls = [],[],[],[],[],[]
dm_t_tt_ls,dm_mt_ls,death_ls,gold_ls,tot_da_ls,tr_dm_ls,win_ls,lane_ls=  [],[],[],[],[],[],[],[]

champions_pd = pd.DataFrame(columns=column)
```

```python
for match in list(data_json.values()):    
    for i in range(10):
        try:
            gameId = match['gameId']
            gameDuration = match['gameDuration']
            championId = match['participants'][i]['championId']
            assists = match['participants'][i]['stats']['assists']
            kills = match['participants'][i]['stats']['kills']
            damageDealtToObjectives = match['participants'][i]['stats']['damageDealtToObjectives']
            damageDealtToTurrets =  match['participants'][i]['stats']['damageDealtToTurrets']
            damageSelfMitigated = match['participants'][i]['stats']['damageSelfMitigated']
            deaths = match['participants'][i]['stats']['deaths'] * -1
            goldEarned = match['participants'][i]['stats']['goldEarned']
            totalDamageDealtToChampions = match['participants'][i]['stats']['totalDamageDealtToChampions']
            trueDamageDealt = match['participants'][i]['stats']['trueDamageDealt']
            if_win = match['participants'][i]['stats']['win']
            lane = match['participants'][i]['timeline']['lane']

            gameids_ls.append(gameId)
            gametime_ls.append(gameDuration)
            champid_ls.append(championId)
            assist_ls.append(assists)
            kill_ls.append(kills)
            dm_t_ob_ls.append(damageDealtToObjectives)
            dm_t_tt_ls.append(damageDealtToTurrets)
            dm_mt_ls.append(damageSelfMitigated)
            death_ls.append(deaths)
            gold_ls.append(goldEarned)
            tot_da_ls.append(totalDamageDealtToChampions)
            tr_dm_ls.append(trueDamageDealt)
            win_ls.append(int(if_win))
            lane_ls.append(lane)
        except:
            continue
```

```python
gameId = match['gameId']
gameDuration = match['gameDuration']
championId = match['participants'][0]['championId']
assists = match['participants'][0]['stats']['assists']
kills = match['participants'][0]['stats']['kills']
damageDealtToObjectives = match['participants'][0]['stats']['damageDealtToObjectives']
damageDealtToTurrets =  match['participants'][0]['stats']['damageDealtToTurrets']
damageSelfMitigated = match['participants'][0]['stats']['damageSelfMitigated']
deaths = match['participants'][0]['stats']['deaths'] * -1
goldEarned = match['participants'][0]['stats']['goldEarned']
totalDamageDealtToChampions = match['participants'][0]['stats']['totalDamageDealtToChampions']
trueDamageDealt = match['participants'][0]['stats']['trueDamageDealt']
if_win = match['participants'][0]['stats']['win']
lane = match['participants'][0]['timeline']['lane']
```

```python
champions_pd.gameId = gameids_ls; champions_pd.gameDuration = gametime_ls
champions_pd.championId = champid_ls; champions_pd.assists = assist_ls
champions_pd.kills = kill_ls; champions_pd.damageDealtToObjectives = dm_t_ob_ls
champions_pd.damageDealtToTurrets = dm_t_tt_ls; champions_pd.damageSelfMitigated = dm_mt_ls
champions_pd.deaths = death_ls;champions_pd.goldEarned = gold_ls
champions_pd.totalDamageDealtToChampions = tot_da_ls;champions_pd.trueDamageDealt = tr_dm_ls
champions_pd.if_win = win_ls;champions_pd.lane = lane_ls
```
